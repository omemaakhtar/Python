```python
import csv 
```

```python
borrowers = [["name", "salary", "credit_score", "debt_ratio", "loan_amount"],
            ["omema", 200000, 650, 0.5, 2500000],
            ["sara", 50000, 750, 0.3, 500000],
            ["farah", 95000, 650, 0.7, 800000],
            ["Ali", 120000, 800, 0.65, 1000000]]
```

```python
with open("borrowers.cvs", "w", newline="")as file:
    writer = csv.writer(file)
    writer.writerows(borrowers)
    
print("CSV File create Successfully")
```

```python
def assess_borrowers(name, salary, credit_score, debt_ratio, loan_amount):
    monthly = (loan_amount * 0.12 * 5 + loan_amount) / (5*12)
    affordability = (monthly / salary *100)
    
    if salary >= 150000 and credit_score >= 600 and debt_ratio < 0.65:
        decision = "Approved"
    elif salary >= 100000 and credit_score >= 500 and debt_ratio < 0.75:
        decision = "Conditional Approval"
    else:
        decision = "Decline"
    
    return decision, round(monthly, 2), round (affordability, 2)
```

```python
with open("borrowers.csv", "r") as infile, open("assessment_results.txt", "w") as outfile:
    reader = csv.reader(infile)
    next(reader)
    
    outfile.write("BATCH LOAN ASSESSMENT REPORT\n")
    outfile.write("="*50 + "\n\n")
    
    for row in reader:
        name = row[0]
        salary = float(row[1])
        credit_score = int(row[2])
        debt_ratio = float(row[3])
        loan_amount = float(row[4])
        
        decision, monthly, affordability = assess_borrower(
            name, salary, credit_score, debt_ratio, loan_amount)
        
        outfile.write(f"Borrower: {name}\n")
        outfile.write(f"Monthly Installment: PKR {monthly:,}\n")
        outfile.write(f"Affordability: {affordability}%\n")
        outfile.write(f"Decision: {decision}\n")
        outfile.write("-"*50 + "\n\n")
        
        print(f"{name}: {decision}")

print("\nReport saved to assessment_results.txt")
```
