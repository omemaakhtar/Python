```python
def create_borrower_profile(name, salary, credit_score, debt_ratio, loan_amount):
    profile = {
        "name" : name,
        "salary": salary,
        "credit_score": credit_score,
        "debt_ratio" : debt_ratio,
        "affordability" : str (round(((loan_amount * 0.12 * 5 + loan_amount)/ (5 *12 )) / salary * 100, 2)) + "%"
    }    
    return profile
            
def print_profile(profile):
    print("\n---Borrower Profile---")
    for key, value in profile.items():
        print(f"{key}, {value}")
        
omema = create_borrower_profile ("omema", 200000, 750, 0.35, 5000000)
ali = create_borrower_profile ("ali", 350000, 685, 0.45, 3500000)

print_profile(omema)
print_profile(ali)
```

```python
def clean_borrower_data (name, city, occupation, cnic):
    cleaned = {
    "name":name.strip().title(),
    "city":city.strip().title(),
    "occupation": occupation.strip().title(),
    "cnic": cnic.strip().replace("-", ""),
    "cnic_region":cnic.strip().split("-")[0]
    }
    return cleaned
def print_cleaned(profile):
    print ("\n---Cleaned Borrower Data")
    for key, value in profile.items():
        print (f"{key}: {value}")

raw1 = clean_borrower_data("omemA", "  KARACHI  ", "BANking", "422201-3366587-8")
raw2 = clean_borrower_data("ALI", "ISlamabad", "teaching", "433512-1234569-8")

print_cleaned(raw1)
print_cleaned(raw2)
```
