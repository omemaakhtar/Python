```python
# Loan Eligibility Calculator

name = (input("Enter Borrower's name:"))
salary = float(input("Enter Borrowers salary in PKR:"))
debt_to_income = float(input("Enter Borrower Debt to Income ratio:"))
credit_score = int(input("Enter Borrower credit score:"))
has_collateral = input("Does Borrower have collateral? (yes/no):")

collateral = has_collateral.lower() =="yes"
print(f"\n--- Credit Assessment for {name}---")

if salary >= 150000 and credit_score >= 600 and debt_to_income < 0.65 and collateral:
    print("Decision, Approved")
    print("Risk Catagoery: Low to Medium")
    
elif salary >=150000 and credit_score >= 500 and debt_to_income < 0.75:
    print("Decision: Approve with conditions")
    print("Risk Catagoery: Medium to High, refer to the credit committee")
    
else: 
    print("Decision: Decline")
    print("Risk Catagoery: Unacceptable")
    
```
