```python
loan_amount = 500000
interest_rate = 0.025
tenure_years = 10

interest = loan_amount*interest_rate*tenure_years
total_repayment = loan_amount + interest
annual_repayment = total_repayment / tenure_years

print("Loan Amount", loan_amount)
print("Interest", interest)
print("Total Repayment", total_repayment)
print("Annual Repayment", annual_repayment)
```

```python
name = input("Enter borrower's name:")
salary =float(input("Enter borrower's salary in PKR:"))
expenses = float(input("Enter borrower's expenses in PKR:"))
saving_goals = float(input("Enter borrower's saving goals in PKR:"))

monthly_saving = salary - expenses
months_to_goals = saving_goals / monthly_saving

print(f"\nBorrower: {name}")
print(f"Monthly Saving: PKR {monthly_saving}")
print(f"Months to reach goals: {months_to_goals}")
print(f"Years to reach goal: {months_to_goals / 12}")
```
