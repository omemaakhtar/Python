```python
borrower_names = ["Omema", "Sadia", "Kashif", "Rehman", "Ali"]
salaries = [200000,500000,600000,350000,80000]
debt_ratio = [0.25,0.3,0.25,0.65,0.7]
credit_scores = [650,675,550,800,750]

for i in range(len(borrower_names)):
    print(f"\n---Assessment for {borrower_names[i]}---")
    if salaries [i] >= 150000 and credit_scores [i] >=600 and debt_ratio [i] < 0.7:
        print("Decision: Approved")
    elif salaries [i]>= 150000 and credit_scores [i]>= 550 and debt_ratio [i] < 0.3:
        print("Approved with conditions: refer to CC")
    else:
        print("Decision: Declined")
```

```python
attempts = 0
maximum_attempts = 3

while attempts < maximum_attempts:
    print(f"loan verification attempt {attempts +1}")
    attempts += 1
print("Maximum attempts reached, Application locked")
```

```python
credit_scores = [650,450,550,700,800]

for scores in credit_scores:
    if scores < 500:
        print(f"Scores {scores} - Below minimum, Skipping")
        continue
    if scores> 700:
        print(f"Scores {scores}- Above avergae - fast approval" )
        break
    print(f"Scores {scores}- Eligible.")
    
#continue — this applicant fails one check, skip to next applicant
# break — found a premium client, no need to keep scanning, act immediately
    
```
