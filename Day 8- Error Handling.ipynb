```python
def get_valid_input (prompt, input_type=float, min_value = 0):
    while True:
        try:
            value = input_type(input(prompt))
            if value <=min_value:
                print(f"Value must be grater than{min_value}. Try again")
                continue
            return value
        except ValueError:
                print("This is invalid entry")
def calculate_loan():
    print("\n---Loan Calculator---")
    try:
        name = input("Borrower Name:")
        salary = get_valid_input("Monthly Salary (PKR):")
        loan_amount = get_valid_input("Loan Amount(PKR):")
        interest_rate = get_valid_input("Annual Interest Rate (e.g. 0.02):")
        tenure = get_valid_input("Tenure in years" , int)
        
        monthly = (loan_amount * interest_rate * tenure + loan_amount) / (tenure * 12)
        affordability = monthly/ salary * 100
        
        print(f"\n---Summary for {name}---")
        print(f" Monthly Installement: PKR{round (monthly, 2):,}")
        print(f"Affordability: {round(affordability, 2)}%")
        
        if affordability > 50:
            print("Warning: High affordability, reveiw carefully")
        else:
            print("Within acceptable range.")
            
    finally:
            print("\nCalculation Session Ended")

calculate_loan()       
        

        
        
        
```
