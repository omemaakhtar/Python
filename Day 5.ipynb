```python
def greet_borrower(name):
    print(f" Welcome {name}. Your application has been processed.")
    
greet_borrower("Sadia")
greet_borrower("Omema")
greet_borrower("kasif")

#def is a function which defines the borrower's name
# (name) is a placeholder which shall replace the name everytime we call the function.
```

```python
def calculate_monthly_installments(loan_amount, interest_rate, loan_tenure):
    total_interest = loan_amount * interest_rate
    total_repayment = loan_amount + total_interest
    monthly_installments = total_repayment / (loan_tenure * 12)
    return monthly_installments

installment = calculate_monthly_installments(500000,0.02,5)
print(f"You monthly installment PKR {round(installment, 2)}")
```

```python
print(calculate_monthly_installments(500000,0.03,10))
print(calculate_monthly_installments(2000000,0.025,15))
print(calculate_monthly_installments(700000,0.04,12))
```

```python
def assess_borrower (name, salaries, debt_ratio, credit_score, loan_amount):
    print(f"\n---Credit Asssessment for {name}---")
    
    installment = calculate_monthly_installments (loan_amount, 0.12, 5)
    affordability = installment / salaries
    
    if salaries >= 100000 and credit_score >700 and debt_ratio < 0.6:
        decision = "Approved"
    elif salaries >= 100000 and credit_score >600 and debt_ratio < 0.55:
        decision = "Approved with Conditions"
    else:
        decision = "Decline"
        
    print(f"Monthly installement: PKR {round(installment, 2)}")
    print(f"Affordability Ratio: {round(affordability * 100, 2)}%")
    print(f"Decision:{decision}")
    
assess_borrower("Omema", 300000, 0.3, 750, 1000000)
assess_borrower("Sadia", 250000, 0.7, 550, 500000)
```

```python
def assess_loan (loan_amount, interest_rate=0.12, tenure_years=5):
    total_interest = loan_amount * interest_rate * tenure_years
    total_repayment = loan_amount + total_interest
    monthly_repayment = total_repayment / (tenure_years *12)
    return round (monthly_repayment, 2)

#Using Defaults
print(assess_loan(500000))

#Overriding defaults
print(assess_loan (500000, interest_rate=0.08))
print(assess_loan(500000, tenure_years= 10))
print(assess_loan(500000, 0.15, 7))
```

```python
def loan_summary (name, loan_amount, interest_rate=0.12, tenure_years = 5):
    total_interest = loan_amount * interest_rate * tenure_years
    total_repayment = loan_amount + total_interest
    monthly = total_repayment / (tenure_years *12)
    
    print(f"---Loan Summary for {name} ---")
    print (f"Loan Amount : PKR {loan_amount:,}")
    print(f"Interest Rate {interest_rate *100}%")
    print(f"Tenure {tenure_years} years")
    print(f"Total Interest: PKR {round(total_interest):,}")
    print (f"Total repayment : PKR {round(total_repayment):,}")
    print (f"Monthly Installment: PKR {round(monthly):,}")
    
loan_summary("Ahmed", 1000000)
loan_summary("Sara", 1000000, 0.12, 7)
loan_summary ("Ali", 500000, interest_rate = 0.08, tenure_years = 8)
          
```
